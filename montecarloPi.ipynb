```python
import numpy.random as rd
```

```python
lenSample = 100000
seed = 9999
rd.seed(seed=seed)
```

```python
countIn = 0
countOut = 0

pi = []

for i in range(lenSample):
  x = rd.uniform(0, 1)
  y = rd.uniform(0, 1)

  if(x**2+y**2 <= 1):
    countIn+=1
  else:
    countOut+=1
  pi.append(4*countIn/lenSample)

totalPoints = countIn + countOut

print(f'''
Total: {totalPoints}                       | {(totalPoints/totalPoints)*100} %
Points inside the circle: {countIn}     | {(countIn/totalPoints)*100} %
Points outside the circle: {countOut}    | {(countOut/totalPoints)*100} %
Pi value: {4*countIn/lenSample}
''')

print('Pi values generated: ', pi)
```
